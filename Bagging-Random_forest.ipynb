```python
#Importinng neccesaary librarys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree 
from sklearn.ensemble import BaggingClassifier    
from sklearn.model_selection import train_test_split , GridSearchCV
from sklearn.metrics import accuracy_score , confusion_matrix , roc_auc_score , roc_curve
from pandas_profiling import ProfileReport
```

```python
df = pd.read_csv('winequality-red.csv')  
```

```python
df.head()
```

```python
x = df.drop(columns=['quality'])
```

```python
y = df.quality
```

```python
#Data spliting
x_train , x_test , y_train , y_test = train_test_split(x,y , random_state  = 0)
```

model_1

```python
bag_dt = BaggingClassifier(DecisionTreeClassifier() , n_estimators=100)
```

```python
bag_dt.fit(x_train  , y_train)
```

```python
bag_dt.predict(x_test)
```

```python
bag_dt.base_estimator_
```

```python
bag_dt.classes_
```

```python
bag_dt.estimator_params
```

model_2

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
bag_knn = BaggingClassifier(KNeighborsClassifier(6) , n_estimators=10)
```

```python
bag_knn.fit(x_train,y_train)
```

```python
bag_knn.predict(x_test)
```

model_3

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=5)
```

```python
rf.fit(x_train, y_train)
```

```python
rf.score(x_test,y_test)
```

```python
bag_dt.score(x_test, y_test)
```

```python
bag_knn.score(x_test,y_test)
```

```python
rf.estimators_
```

```python
rf.estimators_[0]
```

```python
plt.figure(figsize= (20,20))
tree.plot_tree(rf.estimators_[0],filled=True)
plt.show()
```

```python
plt.figure(figsize= (20,20))
tree.plot_tree(rf.estimators_[1],filled=True)
plt.show()
```

```python
plt.figure(figsize = (20,20))
for i in range(len(rf.estimators_)):
    tree.plot_tree(rf.estimators_[i] , filled = True )
```

```python
#auto hyperparameter tester
grid_pram = {
    "n_estimators" : [5,10 , 50 , 100 , 120 , 150],
    'criterion' :['gini' ,'entropy'],
    'max_depth' :range(10),
    'min_samples_leaf' :range(10)
    
}
```

model_4

```python
rf = RandomForestClassifier(n_estimators=5)
```

```python
grid_serach_rf = GridSearchCV(param_grid= grid_pram, cv = 10 , n_jobs=6,verbose=1 ,estimator = rf)
```

```python
grid_serach_rf.fit(x_train,y_train)  #compute havy operation
```

```python
grid_serach_rf.best_params_
```
